## Записка к данным<a class="tocSkip">

Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Вам предстоит построить модель, которая предскажет температуру стали.

### Описание этапа обработки<a class="tocSkip">

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. *slab*, «плита»).

````python
def test_hypothesis(x, y=None, stat='ttest'):
    if stat == 'ttest':
        p = st.ttest_ind(x, y, equal_var=False).pvalue
    elif stat == 'mannwhitneyu':
        p = st.mannwhitneyu(x, y, alternative='two-sided').pvalue
    elif stat == 'shapiro':
        p = st.shapiro(x)[1]
    return p
````

## Процесс обработки стали
[к оглавлению](#start)

Составим таблицу, в которой для каждой партии стали будут собраны все данные о процессе обработки. 

В таблицах с данными о нагреве и измерении температуры одному ключу могут соответствовать сразу несколько строк. Чтобы в итоговой таблице каждому ключу соответствовала единственная строка, нужно для каждого нагрева/измерения создать отдельный столбец. Посмотрим, какого числа столбцов будет достаточно.

Теперь можем создать таблицу. 
* Поскольку наша задача — научиться предсказывать температуру стали, будем рассматривать партии, для которых проводился хотя бы один замер температуры.
* На этапе предварительного анализа было обнаружено, что для некоторых строк в таблице `data_temp` известно время замера, но отсутствует значение температуры. Такие замеры учитывать не будем.

````python
def keep_calm_and_change_everything():
    
    # составим список подходящих ключей
    keys = data_temp.dropna()['key'].unique()
    
    # подготовим таблицу нужного размера
    data_full = pd.DataFrame(
        index=keys, 
        columns= \
            [s + str(i) for i in range(1,17) for s in ['start_heat_','end_heat_','power_active_','power_reactive_']] + \
            [s + str(i) for i in range(1,17) for s in ['time_temp_', 'temp_']] + \
            ['bulk_' + s + str(i) for s in ['volume_', 'time_'] for i in range(1,16)] + \
            ['wire_' + s + str(i) for s in ['volume_', 'time_'] for i in range(1,10)] + ['gas']
    )
    data_full.index.name='key'
    
    # пройдём циклом по всем значениям ключа и построчно заполним таблицу
    for key in notebook.tqdm(data_full.index):
        heat_list = add_nan(
            list(data_arc[data_arc['key'] == key].sort_values(by='start_heat')[data_arc.columns[1:]].values.reshape(-1)),
            n = 16 * 4
        )
        temp_list = add_nan(
            list(data_temp.dropna()[data_temp['key'] == key].sort_values(by='time')[data_temp.columns[1:]].values.reshape(-1)),
            n = 16 * 2
        )
        bulk_list = add_nan(list(data_bulk[data_bulk['key'] == key].values.reshape(-1))[1:], n=15)
        bulk_time_list = add_nan(list(data_bulk_time[data_bulk_time['key'] == key].values.reshape(-1))[1:], n=15)
        wire_list = add_nan(list(data_wire[data_wire['key'] == key].values.reshape(-1))[1:], n = 9)
        wire_time_list = add_nan(list(data_wire_time[data_wire_time['key'] == key].values.reshape(-1))[1:], n=9)
        gas_list = add_nan(list(data_gas[data_gas['key'] == key].values.reshape(-1))[1:], n=1)
        
        data_full.loc[key] = heat_list + temp_list + bulk_list + bulk_time_list + wire_list + wire_time_list + gas_list 
    
    data_full = data_full.reset_index()
    
    # приведём данные к правильным типам
    time_columns = [col for col in data_full.columns if ('start' in col) | ('end' in col) | ('time' in col)] 
    num_columns = [col for col in data_full.columns if (col not in time_columns) & (col != 'key')]
    for col in data_full.columns:
        if col in time_columns:
            data_full[col] = pd.to_datetime(data_full[col].fillna(np.nan), format='%Y-%m-%d %H:%M:%S')
        elif col in num_columns:
            data_full[col] = data_full[col].astype('Float64')
    
    # изменим значения в столбцах с датами на число секунд, прошедших с начала первого нагрева стали
    # точку отсчёта сохраним в столбце `zero_time`
    zero_time = data_full['start_heat_1'].copy()
    for col in time_columns:
        data_full[col] = (data_full[col] - zero_time).apply(lambda t: t.seconds)
    
    return data_full
````

````python
def add_nan(data_list, n):
    length = len(data_list)
    if length == 0:
        return [np.nan] * n
    elif length != n:
        return data_list + [np.nan] * (n - length)
    return data_list
````

Каждой партии стали соответствует единственная строка. Столбцов 146:

# Подготовка выборки

Выборку сохраним в таблице `data`.

````python
data = data_full.copy()
data = data.set_index('key')
````

Оставим только те партии, для которых температура измерялась по крайней мере два раза.

````python
n = len(data)
data = data[~data['temp_2'].isna()]
print('Удалили {} партии'.format(n - len(data)))
````

Добавим столбцы `temperature` и `temperature_init` с начальной и конечной температурами и столбец `process_time` со временем, прошедшим между этими замерами.  
(`n_first, n_last` — вспомогательные столбцы с порядковыми номерами первого и последнего замеров температуры)

````python
def get_temp_data(row):
    # номер первого и последнего замера температуры
    first = str(np.nanargmin(row[time_temp_columns].values) + 1)
    last = str(np.nanargmax(row[time_temp_columns].values) + 1)
    return pd.Series({
        'n_first': first,
        'n_last': last,
        'process_time': row['time_temp_' + last] - row['time_temp_' + first],
        'temperature_init': row['temp_' + first],
        'temperature': row['temp_' + last]
    })
````

````python
temp_meas_columns = ['process_time', 'temperature_init', 'temperature']
data[['n_first', 'n_last'] + temp_meas_columns] = data.apply(get_temp_data, axis=1)
````

Удалим данные о всех нагревах, произошедших после последнего замера температуры, и рассчитаем количество запусков нагрева, их общую длительность и суммарные активную/реактивную мощности. Поскольку нагрев мог производиться до первого замера температуры, столбцов будет восемь:
* `n_heat, heat_time, power_active, power_reactive` — для нагревов после замера начальной температуры
* `pre_n_heat, pre_heat_time, pre_power_active, pre_power_reactive` — для нагревов до замера начальной температуры.

Кажется, что нагрев стали до замера начальной температуры не должен влиять на конечную температуру, но пока оставим данные о нём.

````python
def get_heat_data(row_init):
    row = row_init.copy()
    for i in range(1,17):
        # удаляем поздний нагрев
        if row['start_heat_' + str(i)] > row['time_temp_' + row['n_last']]:
            row.loc[
                [col + str(i) for col in ['start_heat_', 'end_heat_', 'power_active_', 'power_reactive_']]
            ] = np.nan
    # ищем номер последнего раннего нагрева
    n = len([col for col in end_heat_columns if row['time_temp_' + row['n_first']] > row[col]])
    # вычисляем нужные признаки
    row['pre_n_heat'] = n
    row['pre_heat_time'] = row[end_heat_columns[:n]].sum() - row[start_heat_columns[:n]].sum()
    row['pre_power_active'] = row[power_active_columns[:n]].sum()
    row['pre_power_reactive'] = row[power_reactive_columns[:n]].sum()
        
    row['n_heat'] = len([col for col in start_heat_columns if row['time_temp_' + row['n_first']] < row[col]])
    row['heat_time'] = row[end_heat_columns[n:]].sum() - row[start_heat_columns[n:]].sum()
    row['power_active'] = row[power_active_columns[n:]].sum()
    row['power_reactive'] = row[power_reactive_columns[n:]].sum()
        
    return row
````

````python
data = data.apply(get_heat_data, axis=1)
````

Осталось обработать данные о легировании. Удалим все записи о легировании после замера конечной температуры и добавим количество добавленных примесей. Столбцов будет четыре:
* `n_bulk, n_wire` — число примесей, добавленных между замерами начальной и конечной температур
* `pre_n_bulk, pre_n_wire` — число примесей, добавленных перед замером начальной температуры.

````python
def get_alloy_data(row_init):
    row = row_init.copy()  
    
    pre_n_bulk = 0
    for i in range(1,16):
        if row['bulk_time_' + str(i)] > row['time_temp_' + row['n_last']]:
            row.loc[['bulk_time_' + str(i), 'bulk_volume_' + str(i)]] = np.nan
        if row['bulk_time_' + str(i)] < row['time_temp_' + row['n_first']]:
            row.loc[['bulk_time_' + str(i), 'bulk_volume_' + str(i)]] = np.nan
            pre_n_bulk += 1       
    row['pre_n_bulk'] = pre_n_bulk
    row['n_bulk'] = 15 - row[bulk_time_columns].isna().sum()
    
    pre_n_wire = 0
    for i in range(1,10):
        if row['wire_time_' + str(i)] > row['time_temp_' + row['n_last']]:
            row.loc[['wire_time_' + str(i), 'wire_volume_' + str(i)]] = np.nan
        if row['wire_time_' + str(i)] < row['time_temp_' + row['n_first']]:
            row.loc[['wire_time_' + str(i), 'wire_volume_' + str(i)]] = np.nan
            pre_n_wire += 1
    row['pre_n_wire'] = pre_n_wire
    row['n_wire'] = 9 - row[wire_time_columns].isna().sum()  

    return row
````

### Константная модель<a class="tocSkip">
    
Константная модель предскажет на всех объектах тестовой выборки медианную температуру обучающей выборки.

````python
median = y_train.median()
prediction_const =  pd.Series([median] * len(y_test), index=y_test.index)
````

````python
def get_scores(true, pred, time=(0,0)):
    scores = pd.Series({
        'MAE': mean_absolute_error(true, pred),
        'R2': r2_score(true, pred), 
        'RMSE': mean_squared_error(true, pred)**0.5,
        'fit_time': time[0],
        'score_time': time[1]
    })
    return scores
````

Рассмотрим несколько моделей и подберём для них оптимальные гиперпараметры. Оценивать качество будем на кросс-валидации.

````python
def get_best_params(pipeline, parameters_grid, x, y):
    """ Функция для подбора оптимальных гиперпараметров модели по сетке. """
    
    gs = GridSearchCV(pipeline, parameters_grid, cv=3, scoring='neg_mean_absolute_error')
    gs.fit(x, y)
    print('MAE (кросс-валидация): {:.3f}'.format(-gs.best_score_))
    return gs.best_params_
````

````python
def get_cv_scores(pipeline, x, y): 
    """ Функция для оценки на кросс-валидации качества модели. """
    
    scoring = ('neg_mean_absolute_error','r2','neg_mean_squared_error')
    cv_results = cross_validate(pipeline, x, y, cv=3, scoring=scoring)
    scores = pd.Series({
        'MAE': -cv_results['test_neg_mean_absolute_error'].mean(),
        'R2': cv_results['test_r2'].mean(),
        'RMSE': -cv_results['test_neg_mean_squared_error'].mean(),
        'fit_time': cv_results['fit_time'].mean(),
        'score_time': cv_results['score_time'].mean()
    })
    return scores
````

````python
def add_objects(row, key):
    df = pd.DataFrame(columns=row.index)
    for i in range(1, 16 - row[time_temp_columns].isna().sum()):
        new_row = row.copy()
        new_row[['temp_' + str(j) for j in range(1, i)]] = np.nan
        new_row[['time_temp_' + str(j) for j in range(1, i)]] = np.nan
        df.loc[str(key) + '_' + str(i)] = new_row 
    return df
````
